```python
import matplotlib.pyplot as plt
import numpy as np
import math 
```

```python
#problem 1

def function():
    for n in range(2,21):
        c = n*np.log(n)-n;
        print(c)

        
def function2():
    for n in range(2,21):
        d = np.math.factorial(n);
        print(d)


print(function())
print(function2())

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(c,n,label='Stirlings approximation')
ax.plot(d,n,label='Actual values of factorial')
ax.legend();
```

```python
#problem 5a

arr3 = np.random.randn(1000,2)

x = arr3[:,0]

y = arr3[:,1]

np.random.seed(0)
```

```python
#problem 5b




fig, axes = plt.subplots(1,2)


axes[0].plot(x)
axes[1].plot(y)
axes[0].set_title('Random X values')
axes[0].set_ylabel('x')
axes[1].set_title('Random Y values')
axes[1].set_ylabel('y')
plt.tight_layout();
```

```python
#problem 5c

fig = plt.figure()

ax = fig.add_axes([1,1,1,1])

ax.scatter(x, y, marker='o')
ax.set_xlabel('x')
ax.set_ylabel('y ')
ax.set_title('RandScat');

```

```python
#problem 5d


fig = plt.subplots(figsize =(10, 7))
plt.hist2d(x, y)
plt.title("Random value 2d histogram")
  
plt.show()
```
